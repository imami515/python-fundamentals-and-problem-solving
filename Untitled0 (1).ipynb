# **Checking Palindrome**

```python
def check_palindrome(mystr):
    cleaned_str = mystr.lower()
    reversed_str = cleaned_str[::-1]

    if cleaned_str == reversed_str:
        return True
    else:
        return False

word1 = "Radar"
if check_palindrome(word1):
    print(f"'{word1}' it is Palindrome ")
else:
    print(f"'{word1}' it is not  Palindrome")

word2 = "Python"
if check_palindrome(word2):
    print(f"'{word2}' it is palindrome")
else:
    print(f"'{word2}' it is not palindromr")
```

# **Character Frequency Count(with function)**

```python
def count_characters(text):
    count_dict = {}

    for char in text:
        if char in count_dict:
            count_dict[char] = count_dict[char] + 1
        else:
            count_dict[char] = 1

    return count_dict

string1 = "hello"
result1 = count_characters(string1)
print(f"'{string1}'  count:", result1)

string2 = "apple"
result2 = count_characters(string2)
print(f"'{string2}'  count:", result2)
```

# **String Encryption (Caesar Cipher)**

```python

text = "abc"
shift = 3
encrypted_text = ""

for char in text:
   ascii_value = ord(char)
   new_ascii_value = ascii_value + shift
   new_char = chr(new_ascii_value)
   encrypted_text = encrypted_text + new_char

print("Original:", text)
print("Encrypted:", encrypted_text)
```

# **First Non-Repeating Character**

```python

word = "swiss"
answer = "Koi unique character nahi mila"

for char in word:
    if word.count(char) == 1:
        answer = char
        break

print("first non repeating character :", answer)
```

# **Reverse Words in a Sentence**

```python

sentence = "Python is fun"
words_list = sentence.split()
reversed_list = words_list[::-1]
final_sentence = " ".join(reversed_list)

print("Original:", sentence)
print("Reversed Words:", final_sentence)
```
