# **Combining Two Functions Into One**

```python

def compose(f, g):
    return lambda x: f(g(x))

add_five = lambda x: x + 5
square = lambda x: x ** 2

add_and_square = compose(square, add_five)
print(add_and_square(3))
```

# **Making Slow Code Fast Using Cache**

```python
def memoize(fn):
    cache = {}
    def wrapper(x):
        if x not in cache:
            cache[x] = fn(x)
        return cache[x]
    return wrapper


import time
def expensive_calculation(n):
    time.sleep(1)
    return n * 2

smart_calc = memoize(expensive_calculation)
print(smart_calc(10))
print(smart_calc(10))
```

# **Finding Two Numbers That Equal a Target**

```python
numbers = [1, 2, 4, 6, 8, 11, 15]
target = 14

left = 0
right = len(numbers) - 1
found = False

while left < right:
    current_sum = numbers[left] + numbers[right]
    if current_sum == target:
        print(f"Pair found: {numbers[left]} and {numbers[right]}")
        found = True
        break
    elif current_sum < target:
        left += 1
    else:
        right -= 1

if not found:
    print("No pair exists.")
```


# **Finding the Highest Sum in a Moving Window**

```python
arr = [2, 1, 5, 1, 3, 2]
k = 3

window_sum = sum(arr[:k])
max_sum = window_sum

for i in range(len(arr) - k):
   window_sum = window_sum - arr[i] + arr[i + k]
   if window_sum > max_sum:
      max_sum = window_sum

print(f"Maximum sum of a subarray of size {k} is: {max_sum}")
```

# **Finding the First Letter That Does Not Repeat**

```python
text = "swiss"
char_counts = {}

for char in text:
    char_counts[char] = char_counts.get(char, 0) + 1

first_unique = None
for char in text:
    if char_counts[char] == 1:
        first_unique = char
        break

if first_unique:
    print(f"The first non-repeating character is: '{first_unique}'") # Output: 'w'
else:
    print("All characters repeat.")
```
